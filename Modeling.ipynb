```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pycaret.classification import *
```

# Model with data leaking

```python
df1 = pd.read_csv('0.100maps.csv')
df2 = pd.read_csv('100.300maps.csv')
df3 = pd.read_csv('300.600maps.csv')
df4 = pd.read_csv('300.600maps.csv')
df5 = pd.read_csv('600.900maps.csv')

dfs = [df1,df2,df3,df4,df5]

df = pd.concat(dfs)
df_e = pd.read_csv('economy_df.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['winner'] = df.apply(lambda x :1 if x['score_t1']>x['score_t2'] else 2, axis=1)
```

```python
df['match_id']
```

```python
df_e
```

```python
maps_to_num = dict(zip(df['map'].unique(), range(len(df['map'].unique()))))

df['map'] = df['map'].apply(lambda x : maps_to_num[f'{x}'] if x in maps_to_num else '')

```

```python
df
```

```python
set_of_teams = set(list(df['team_1']) + list(df['team_2']))
team_to_num = dict(zip(set_of_teams, range(len(set_of_teams))))

#df['team_1'] = df['team_1'].replace(team_to_num)
df['team_1'] = df['team_1'].apply(lambda x : team_to_num[f'{x}'] if x in team_to_num else '')
df['team_2'] = df['team_2'].apply(lambda x : team_to_num[f'{x}'] if x in team_to_num else '')
```

```python
df.drop(columns=['score_t1','ct_side_t1','t_side_t1','score_t2','ct_side_t2','t_side_t2','team_1_rating','team_2_rating'], inplace=True)
#df.drop(columns = ['Unnamed: 0','nummap'], inplace=True)
df
```

```python
from pycaret.internal.tabular import LabelEncoder


classification = setup(data = df, target = 'winner', session_id=42)
```

```python
bestmodel  = compare_models()
```

```python
evaluate_model(bestmodel)
```
